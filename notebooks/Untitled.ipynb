```python
data = df['star_distance']

data.describe()
```

```python
import matplotlib.pyplot as plt

fig1, ax = plt.subplots()
ax.set_title('Plot')

ax.boxplot(data, vert=False)
```

```python
import seaborn as sns
plt.figure(figsize=(20,5))
g = sns.boxplot(x = data, palette="Set2")
g.set_xticklabels(g.get_xticks(), size = 15)
g.axes.set_title("Boxplot of distances to exoplanet's star(s)",fontsize=30)
g.set_xlabel("Distance",fontsize=20);
```
